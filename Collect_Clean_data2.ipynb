```python
import pandas as pd
```

# Vaccination Data Columns Definition
https://data.cdc.gov/Vaccinations/COVID-19-Vaccinations-in-the-United-States-Jurisdi/unsk-b7fc

```python
# Vaccination Case USA
Vacc_Case_df = pd.read_csv('COVID-19_Vaccinations_in_the_United_States_Jurisdiction.csv')
Vacc_Case_df.head(5)
```

# Casae & Death Data Columns Definition
https://data.cdc.gov/Case-Surveillance/United-States-COVID-19-Cases-and-Deaths-by-State-o/9mfq-cb36

```python
# Covid Cases and Deaths by State over time
Case_Death_df = pd.read_csv('United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv')
Case_Death_df.head(5)
```

```python
# 5 States data filtered from Vaccination Data
Selected_States_Vacc_df = Vacc_Case_df.loc[(Vacc_Case_df["Location"] == "CA") |
                                 (Vacc_Case_df["Location"] == "TX") |
                                 (Vacc_Case_df["Location"] == "FL") |
                                 (Vacc_Case_df["Location"] == "CA") |
                                 (Vacc_Case_df["Location"] == "MO")
                                 ]
Selected_States_Vacc_df.sort_values(['Date'])
Selected_States_Vacc_df.head(5)
```

```python
# Check Data Types
Selected_States_Vacc_df.dtypes
```

```python
# Check any missing value
Selected_States_Vacc_df.isnull().sum()
```

```python
# 5 States data filtered from Case & Death Data
Selected_States_Case_Death_df = Case_Death_df.loc[(Case_Death_df["state"] == "CA") |
                                 (Case_Death_df["state"] == "TX") |
                                 (Case_Death_df["state"] == "FL") |
                                 (Case_Death_df["state"] == "CA") |
                                 (Case_Death_df["state"] == "MO")
                                 ]
Selected_States_Case_Death_df.sort_values(['submission_date'])
Selected_States_Case_Death_df.head(5)
```

```python
# Check Data Types
Selected_States_Case_Death_df.dtypes
```

```python
# Check any missing value
Selected_States_Case_Death_df.isnull().sum()
```
